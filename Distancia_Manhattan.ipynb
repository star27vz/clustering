```python
import matplotlib.pyplot as plt

x = [185, 170, 168, 179, 182, 188]
y = [72, 56, 60, 68, 72, 77]

# Inicializando centroides
k1 = [185, 72]
k2 = [170, 56]

# Almacenando los puntos en los clustering
cluster_1 = [[185, 72]]
cluster_2 = [[170, 56]]

def distancia_euclidiana(punto, centroide):
    return (abs(punto[0] - centroide[0]) + abs(punto[1] - centroide[1]))

def actualizar_centroide(punto, centroide):
    print("Centroide Actualizado : ",(punto[0] + centroide[0]) / 2,(punto[1] + centroide[1]) / 2)
    return [(punto[0] + centroide[0]) / 2, (punto[1] + centroide[1]) / 2]

# Agrupacin
for i in range(2, len(x)):
    punto = [x[i], y[i]]
    print("Punto :",punto)

    # Calculo de Ditancia Punto - Centroide
    distancia_k1 = distancia_euclidiana(punto, k1)
    print("Distancia Cluster 1: " ,distancia_k1)
    distancia_k2 = distancia_euclidiana(punto, k2)
    print("Distancia Cluster 2: " , distancia_k2)

    # Asignando el punto al centroide ms cercano
    if distancia_k1 < distancia_k2:
        clustering = 1
        cluster_1.append(punto)
    else:
        clustering = 2
        cluster_2.append(punto)

    # Actualizacin de centroides
    if clustering == 1:
        k1 = actualizar_centroide(punto, k1)
    else:
        k2 = actualizar_centroide(punto, k2)


    print("Cluster 1 = ",cluster_1)
    print("Cluster 2 = ",cluster_2)
    plt.figure()
    plt.scatter(*zip(*cluster_1), label="Cluster 1", c='purple')
    plt.scatter(*zip(*cluster_2), label="Cluster 2", c='fuchsia')
    plt.scatter(*k1, marker='x', s=100, c='green', label="Centroide 1")
    plt.scatter(*k2, marker='x', s=100, c='black', label="Centroide 2")
    plt.title(f'Iteracin {i}')
    plt.legend()
    plt.show()
```
